```python
symbols = "A,B,C,D,E,F,G,H,I,J,K,L,M,N"
other_symbols = "O,P,Q,R,S,T,U,V,W,X,Y"

symbols = symbols.split(",")
other_symbols = other_symbols.split(",")
```

```python
import random
n_msgs = 30
msgs = []
for i in range(n_msgs):
    length = random.randint(5, 21)
    cur_msg = ''
    for j in range(length):
        cur_msg += random.choice(symbols)
    msgs.append(cur_msg)
        
    
```

```python
from collections import Counter
c = Counter()
for msg in msgs:
    msg_len = len(msg)
    for i in range(msg_len-1):
        j = i+1
        byte_pair = msg[i:j+1]
        byte_pairs.append(byte_pair)
```

```python
from collections import Counter
c = Counter(byte_pairs)
```

```python
c.most_common()
```

```python
n_other = len(other_symbols)
new_vocab ={}
mc = c.most_common()
for i in range(n_other):
    new_vocab[mc[i][0]] = other_symbols[i]
    
```

```python
new_vocab
```

```python
new_msgs = []
for msg in msgs:
    new_msg = msg
    for nk, nv in new_vocab.items():
        new_msg = new_msg.replace(nk, nv)
    new_msgs.append(new_msg)
```
